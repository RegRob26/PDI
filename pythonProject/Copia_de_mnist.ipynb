#generacion de un conunto con 2 tipos

dig1, dig0



```python
dig1 = 1
dig0 = 0
```

```python

xtemp = train_x.reshape((-1,train_x.shape[1]*train_x.shape[2]))
xtemp.shape
```

```python
X = None #matriz de dos dimensiones
Y = None
bandera = True
for indice in train_y :
  if train_y[indice] == dig0 or train_y[indice] == dig1 :
    if bandera :
      X = xtemp[indice].reshape((-1,train_x.shape[1]* train_x.shape[2]))
      Y = train_y[indice].reshape((-1,1))
      bandera = False
      print(X.shape)
    else :
      X = np.concatenate((X,xtemp[indice].reshape((-1,train_x.shape[1]* train_x.shape[2]))), axis=0)
      Y = np.concatenate((Y,train_y[indice].reshape((-1,1))), axis=0)

```

#Matriz de pesos


```python
np.random.seed(0)
```

```python
W = np.random.random((X.shape[1],1)) * 0.2 - 0.1
print(W.shape)
```

```python
W[:5]
```

#Evaluando el perceptron con los pesos inciales

```python
#definir la funcion signo

def signo(x):
    return (x >= 0) * 2-1
```

```python
Yr  = (Y == 0 ) * 2-1  #respuesta de Y , Yr contiene valores de +1  o -1
```

```python
#indices aleatorios
indice = np.arange(0,X.shape[0])
np.random.shuffle(indice)
```

```python
indice
```

```python
alpha = 0.01

print((signo(X.dot(W).reshape((-1)) == Yr.reshape((-1))).mean()))

for i in indice[:5]:
  X1  = X [i,:].reshape((1,X.shape[1]))
  Yobtenida = signo(X1.dot(W))
  W =  W + alpha * (Yr[i] - Yobtenida) * X1.T

print((signo(X.dot(W).reshape((-1)) == Yr.reshape((-1))).mean()))
```

```python
fig, axs = plt.subplots(4, 4, figsize=(15, 15))

#sobre cada par de características y crear un scatter plot en cada subplot

for i in range(4):
  for j in range(4):
       img = X[i*4+j,1:].reshape((28,28))
       axs[i,j].imshow(img, cmap='gray')
       etiqueta = signo(X[i*4+j,:].dot(W))
       if etiqueta == 1 :
         axs[i,j].set_xlabel("Etiqueta {}".format(dig0))
       else :
         axs[i,j].set_xlabel("Etiqueta {}".format(dig1))

# Ajustar el espaciado entre los subplots
plt.tight_layout()
```
